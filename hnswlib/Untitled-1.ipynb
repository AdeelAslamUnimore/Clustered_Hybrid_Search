```python
import pandas as pd
import random

# Load the dataset from a CSV file
# Replace 'your_file.csv' with the actual file path
df = pd.read_csv('/home/u6059148/DataSet/documents_full.csv', sep=";")
```

```python


# Step 1: Sample 1000 queries from the dataset
sampled_df = df.sample(n=1000, random_state=42).reset_index(drop=True)

# Step 2: Add a new column 'EndDate'
# For each sampled row, randomly select an 'EndDate' from the dataset to ensure it's in range

# Function to assign a random end date based on 'DateCreated' range
def assign_end_date(start_date, df):
    # Filter out rows that have the same or later 'DateCreated' than the start date
    later_dates = df[df['DateCreated'] >= start_date]['DateCreated'].unique()
    if len(later_dates) > 0:
        return random.choice(later_dates)
    else:
        # If no later dates, return the same date
        return start_date

# Apply the function to each row in the sampled dataframe
sampled_df['EndDate'] = sampled_df['DateCreated'].apply(lambda start_date: assign_end_date(start_date, df))

# Step 3: Save the sampled data with the new 'EndDate' column to a CSV file
sampled_df.to_csv('/home/u6059148/DataSet/Range_date_queries.csv', index=False, sep=';')

sampled_df
```

```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('/home/u6059148/Result/TwoHopSearch/100EPS/Q2.csv')

# Display the contents of the CSV file
df
```

```python
# Step 1: Sample the start and end dates from the dataset
# Step 1: Convert 'DateCreated' column to datetime for comparison
df['DateCreated'] = pd.to_datetime(df['DateCreated'])

# Step 2: Sample 1000 rows from the dataset
sampled_df = df.sample(n=1000, replace=True)  # Set 'replace' to False if no duplicates are needed

# Step 3: For each row, assign a random 'StartDate' and 'EndDate'
start_dates = []
end_dates = []

for _ in range(len(sampled_df)):
    start_date = random.choice(df['DateCreated'])
    end_date = random.choice(df['DateCreated'])
    
    # Ensure start_date is earlier than end_date
    if start_date > end_date:
        start_date, end_date = end_date, start_date
    
    start_dates.append(start_date.strftime('%Y-%m-%d'))
    end_dates.append(end_date.strftime('%Y-%m-%d'))

# Step 4: Add StartDate and EndDate columns to the sampled dataset
sampled_df['StartDate'] = start_dates
sampled_df['EndDate'] = end_dates

# Step 5: Select only the necessary columns (without 'DateCreated')
sampled_df = sampled_df[['StartDate', 'EndDate', 'ClinicalAreas', 'embedding']]
sampled_df.to_csv('/home/u6059148/DataSet/Range_date_queries.csv',sep=';', index=False)
```

```python
import os
import pandas as pd
ground_truth_folder = '/home/u6059148/DataSet/GroundTruth/'

algorithm_folder = '/home/u6059148/Result/TwoHopSearch//300EPS'
results = []
# List all files in the ground truth folder
ground_truth_files = os.listdir(ground_truth_folder)

# Iterate through each file
for filename in ground_truth_files:
    # Construct the full file path
    ground_truth_file = os.path.join(ground_truth_folder, filename)
    algorithm_file = os.path.join(algorithm_folder, filename)
    if os.path.exists(algorithm_file):
        # Read the CSV files
        ground_truth_df = pd.read_csv(ground_truth_file)
        algorithm_df = pd.read_csv(algorithm_file)
        algorithm_df = algorithm_df.drop_duplicates()
        
        # Get the first 15 records of the ground truth file (assuming there's at least 15 records)
        ground_truth_first_15 = ground_truth_df.head(10)
        
        match_count = 0
        for index, algorithm_row in algorithm_df.iterrows():
            # Check if the record in the algorithm file matches any record in the first 15 records of ground truth in column A
            if algorithm_row['ID'] in ground_truth_first_15['ID'].values:
                match_count += 1
        results.append({'query': filename, 'match': match_count})
    else:
        print(f"{filename}: Corresponding file not found in the algorithm folder.")
    # Check if the corresponding file exists in the algorithm folder
results_df = pd.DataFrame(results)
results_df.to_csv('/home/u6059148/Result/TwoHopSearch/Res300.csv', index=False)

print("Results have been written to output_matches.csv") 
```

```python
import os
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

ground_truth_folder = '/home/u6059148/DataSet/GroundTruth/'
algorithm_folder = '/home/u6059148/Result/TwoHopSearch//300EPS'
results = []

# List all files in the ground truth folder
ground_truth_files = os.listdir(ground_truth_folder)

# Function to process each file pair and count matches
def process_file(filename):
    ground_truth_file = os.path.join(ground_truth_folder, filename)
    algorithm_file = os.path.join(algorithm_folder, filename)
    if os.path.exists(algorithm_file):
        # Read the CSV files
        ground_truth_df = pd.read_csv(ground_truth_file)
        algorithm_df = pd.read_csv(algorithm_file)
        algorithm_df = algorithm_df.drop_duplicates()

        # Get the first 10 records of the ground truth file
        ground_truth_first_15 = ground_truth_df.head(10)

        match_count = 0
        for index, algorithm_row in algorithm_df.iterrows():
            # Check if the record in the algorithm file matches any record in the first 10 records of ground truth in column 'ID'
            if algorithm_row['ID'] in ground_truth_first_15['ID'].values:
                match_count += 1

        return {'query': filename, 'match': match_count}
    else:
        print(f"{filename}: Corresponding file not found in the algorithm folder.")
        return None

# Using ThreadPoolExecutor for parallel processing
with ThreadPoolExecutor() as executor:
    # Map the filenames to the process_file function and collect results
    results = list(filter(None, executor.map(process_file, ground_truth_files)))

# Save results to CSV
results_df = pd.DataFrame(results)
results_df.to_csv('/home/u6059148/Result/TwoHopSearch/Res300.csv', index=False)

print("Results have been written to Res200.csv")
```

```python
import os
import pandas as pd
from concurrent.futures import ProcessPoolExecutor

ground_truth_folder = '/home/u6059148/DataSet/GroundTruth/'
algorithm_folder = '/home/u6059148/Result/PostFiltering'
results = []

# List all files in the ground truth folder
ground_truth_files = os.listdir(ground_truth_folder)

# Function to process each file pair and count matches
def process_file(filename):
    ground_truth_file = os.path.join(ground_truth_folder, filename)
    algorithm_file = os.path.join(algorithm_folder, filename)
    
    if os.path.exists(algorithm_file):
        # Read only the 'ID' column for faster processing
        ground_truth_df = pd.read_csv(ground_truth_file, usecols=['ID']).head(10)  # First 10 records
        algorithm_df = pd.read_csv(algorithm_file, usecols=['ID']).drop_duplicates()

        # Use vectorized matching with isin
        match_count = algorithm_df['ID'].isin(ground_truth_df['ID']).sum()

        return {'query': filename, 'match': match_count}
    else:
        print(f"{filename}: Corresponding file not found in the algorithm folder.")
        return None

# Using ProcessPoolExecutor for parallel processing
with ProcessPoolExecutor() as executor:
    # Map the filenames to the process_file function and collect results
    results = list(filter(None, executor.map(process_file, ground_truth_files)))

# Save results to CSV
results_df = pd.DataFrame(results)
results_df.to_csv('/home/u6059148/Result/PostFiltering/Res500.csv', index=False)

print("Results have been written to Res400_2.csv")
```

```python
import pandas as pd
import glob
import os

# Path where your files are located
file_path = "/home/u6059148/Result/PostFiltering/"
all_files = glob.glob(os.path.join(file_path, "Q*.csv"))  # Matches Q0.csv, Q1.csv, ..., Q999.csv

# List to hold DataFrames from each file
dfs = []

# Iterate through each file and read it into a DataFrame
for filename in all_files:
    df = pd.read_csv(filename)
    dfs.append(df)

# Concatenate all DataFrames into one
merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_file_path = "/home/u6059148/Result/PostFiltering/merged_results.csv"
merged_df.to_csv(merged_file_path, index=False)

print(f"Merged file saved as {merged_file_path}")
```
