```python
import pandas as pd
import numpy as np
import time
from multiprocessing import Pool, cpu_count

# -----------------------------
# CONFIGURATION
# -----------------------------
N = len(df_docs)            # Number of documents
selectivity = 0.10          # 10%
target = int(selectivity * N)
batch_size = 50             # queries per batch
output_file = "doc_query_attr_2000.csv"

df_docs = df_docs.reset_index(drop=True)
# Only first 2000 queries
queries = df_queries["query_attribute"].astype(str).values[:2000]

# -----------------------------
# HELPER FUNCTION FOR PARALLEL
# -----------------------------
def process_query(q):
    """Assign a single query to random documents"""
    chosen = np.random.choice(N, size=target, replace=False)
    return [(idx, q) for idx in chosen]

# -----------------------------
# MAIN PARALLEL BATCH LOOP
# -----------------------------
start_time = time.time()
first_write = True

for batch_start in range(0, len(queries), batch_size):
    batch_end = min(batch_start + batch_size, len(queries))
    batch_queries = queries[batch_start:batch_end]

    # Parallel processing
    with Pool(processes=cpu_count()) as pool:
        results = pool.map(process_query, batch_queries)

    # Flatten results into doc -> list of queries
    doc_attr = [[] for _ in range(N)]
    for res in results:
        for idx, q in res:
            doc_attr[idx].append(q)

    # Convert lists to comma-separated strings
    doc_attr_str = [",".join(x) for x in doc_attr]

    # Save batch to CSV
    df_out = pd.DataFrame({
        "doc_id": df_docs.index,
        "query_attribute": doc_attr_str
    })
    df_out.to_csv(output_file, mode="w" if first_write else "a",
                  index=False, header=first_write)
    first_write = False

    # Progress display
    elapsed = time.time() - start_time
    processed = batch_end
    percent = (processed / len(queries)) * 100
    print(f"Processed {processed}/{len(queries)} queries "
          f"({percent:.2f}%) | Time elapsed: {elapsed/60:.2f} min")

print("✅ Finished writing comma-separated attributes for first 2000 queries.")
```

```python
import random
from collections import Counter
import pandas as pd

TOTAL = 10000
data = []

# Fixed counts for 1–4
data.extend([1] * 2950)
data.extend([2] * 1972)
data.extend([3] * 1479)
data.extend([4] * 986)

# Remaining rows for 5–12
remaining = TOTAL - len(data)  # 2,613
values_5_to_12 = list(range(5, 13))  # 8 numbers
per_value = remaining // len(values_5_to_12)  # 326
extra = remaining % len(values_5_to_12)       # 5 leftover rows

for i, v in enumerate(values_5_to_12):
    count = per_value + (1 if i < extra else 0)  # distribute leftover
    data.extend([v] * count)

# Shuffle
random.shuffle(data)

# Load into pandas
df_random_items = pd.DataFrame(data, columns=["random_value"])

# Verification
print("Total rows:", len(df_random_items))  # Should be 10,000
print(Counter(df_random_items["random_value"]))
```

```python
import os
import pandas as pd

# Example: load your dataframe
# df = pd.read_csv("your_dataset.csv")
# Make sure index matches those numbers
# df.index = df.index.astype(int)
df = pd.read_csv("/scratch/aa5f25/datasets/TripClick/documents_full.csv", sep=";")


folder_path = "/scratch/aa5f25/datasets/TripClick/Corelation"  # <-- your folder path

# Create empty column
df["correlation_files"] = ""

for file in os.listdir(folder_path):
    if not file.endswith(".csv"):
        continue

    file_path = os.path.join(folder_path, file)

    # Read the single-line file
    with open(file_path, "r") as f:
        line = f.readline().strip()

    if not line:
        continue

    # Split left and right
    key, idx_string = line.split(";")

    # Convert index positions to list of integers
    indices = list(map(int, idx_string.split(",")))

    # Keep only valid indices (avoid crash if bad index exists)
    valid_indices = [i for i in indices if i in df.index]

    # Update dataframe
    for idx in valid_indices:
        if df.at[idx, "correlation_files"] == "":
            df.at[idx, "correlation_files"] = key
        else:
            df.at[idx, "correlation_files"] += f",{key}"

print("Done processing all files.")
```
